## 1. Setup & Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

plt.rcParams['figure.figsize'] = (12, 5)
sns.set_theme(style='whitegrid')
pd.set_option('display.max_columns', None)
```

## 2. Load Data

```python
from datasets import load_dataset

# Load primary dataset from HuggingFace
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

print(f'Shape: {df.shape}')
df.head()
```

```python
# Load BLS OEWS enrichment dataset
# Download from: https://www.bls.gov/oes/tables.htm
# Place the file in data/raw/ and update the path below

# oews = pd.read_excel('data/raw/oews_national.xlsx', header=1)
# print(f'OEWS shape: {oews.shape}')
# oews.head()
```

## 3. Data Cleaning

```python
# Check missing values
print('Missing values:')
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Filter rows with salary data
df_salary = df.dropna(subset=['salary_year_avg']).copy()
print(f'Rows with salary data: {len(df_salary)} / {len(df)}')
```

## 4. Exploratory Data Analysis (EDA)

```python
# Salary distribution
plt.figure(figsize=(12, 5))
sns.histplot(df_salary['salary_year_avg'], bins=50, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Annual Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
# Top 10 job titles by count
top_titles = df_salary['job_title_short'].value_counts().head(10)
sns.barplot(x=top_titles.values, y=top_titles.index)
plt.title('Top 10 Job Titles')
plt.xlabel('Count')
plt.tight_layout()
plt.show()
```

## 5. Hypothesis Tests

```python
# H1: Remote jobs pay more than on-site jobs
remote = df_salary[df_salary['job_work_from_home'] == True]['salary_year_avg']
onsite = df_salary[df_salary['job_work_from_home'] == False]['salary_year_avg']

stat, p_value = stats.mannwhitneyu(remote, onsite, alternative='greater')
print(f'Remote median: ${remote.median():,.0f}')
print(f'On-site median: ${onsite.median():,.0f}')
print(f'Mann-Whitney U stat: {stat:.2f}, p-value: {p_value:.4f}')
print('Result:', 'Reject H0 (remote pays more)' if p_value < 0.05 else 'Fail to reject H0')
```
